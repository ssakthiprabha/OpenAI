```python
import json
from PIL import Image
import pytesseract

# Set the path to the Tesseract executable
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Open the image file
image = Image.open(r'C:\Users\ssathiyaseelan\Desktop\Sakthi\python\OCR\Image\3994267-1.jpg')

# Perform OCR
raw_text = pytesseract.image_to_string(image)

# Format the extracted text for readability
formatted_text = raw_text.strip().replace('\n', ' ').replace('  ', ' ')

# Create a Python dictionary with an easily understandable response
response = {
    "status": "success",
    "message": "Text successfully extracted from the image.",
    "data": {
        "text_summary": formatted_text[:100] + "..." if len(formatted_text) > 100 else formatted_text,
        "full_text": formatted_text
    }
}

# Convert the dictionary to a JSON-formatted string
json_response = json.dumps(response, indent=4)

# Print the JSON response
print(json_response)
```

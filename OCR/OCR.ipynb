```python
import json
import pandas as pd
from PIL import Image
import pytesseract

# Set the path to the Tesseract executable
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Open the image file
image = Image.open(r'C:\Users\ssathiyaseelan\Desktop\Sakthi\python\OCR\Image\3994267-1.jpg')

# Perform OCR
raw_text = pytesseract.image_to_string(image)

# Process the text to extract a table format
rows = raw_text.splitlines()
table_data = [row.split() for row in rows]  # Split by whitespace to form rows and columns

# Convert the list into a Pandas DataFrame
df = pd.DataFrame(table_data)

# Convert DataFrame to a JSON-friendly format
table_json = df.to_json(orient='records', lines=True)

# Create a Python dictionary for the response
response = {
    "status": "success",
    "message": "Table successfully extracted from the image.",
    "data": {
        "table": table_json
    }
}

# Convert the dictionary to a JSON-formatted string
json_response = json.dumps(response, indent=4)

# Print the JSON response
print(json_response)
```

```python
import json
from PIL import Image
import pytesseract

# Set the path to the Tesseract executable
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Open the image file
image = Image.open(r'C:\Users\ssathiyaseelan\Desktop\Sakthi\python\OCR\Image\3994267-1.jpg')

# Perform OCR to extract the text
raw_text = pytesseract.image_to_string(image)

# Use Tesseract's layout analysis to extract detailed information
data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

# Initialize variables to store relevant data
sales_rep_info = []
sales_rep_found = False

# Look for the term 'Sales Rep' in the extracted text and extract relevant data
for i, text in enumerate(data['text']):
    if 'BILL TO' in text.lower():  # Case-insensitive match for 'sales rep'
        sales_rep_info.append({
            "text": text,
            "top": data['top'][i],
            "left": data['left'][i],
            "width": data['width'][i],
            "height": data['height'][i]
        })
        sales_rep_found = True

# If sales rep information is found, return it as a structured JSON response
if sales_rep_found:
    response = {
        "status": "success",
        "message": "Sales Rep information successfully extracted.",
        "data": {
            "sales_rep_info": sales_rep_info
        }
    }
else:
    response = {
        "status": "error",
        "message": "Sales Rep information not found.",
        "data": {}
    }

# Convert the response to a JSON-formatted string
json_response = json.dumps(response, indent=4)

# Print the JSON response
print(json_response)
```
